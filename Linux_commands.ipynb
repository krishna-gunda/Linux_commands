# Basic Linux Commands

## 1. `ls` – List Files and Directories

Displays files and folders in the current directory.

```bash
ls
```

---

## 2. `pwd` – Print Current Directory

Shows the current working directory path.

```bash
pwd
```

---

## 3. `cd` – Change Directory

Move from one directory to another.

```bash
cd folder_name
```

Example:

```bash
cd Documents
```

---

## 4. `mkdir` – Create New Directory

Creates a new folder.

```bash
mkdir test
```

---

## 5. `rmdir` – Remove Empty Directory

Deletes an empty folder.

```bash
rmdir test
```

---

# File Handling Commands

## 6. `touch` – Create File

Creates an empty file.

```bash
touch file.txt
```

---

## 7. `cat` – Display File Content

Shows file content in terminal.

```bash
cat file.txt
```

---

## 8. `cp` – Copy Files

Copies files from one location to another.

```bash
cp file.txt backup/
```

---

## 9. `mv` – Move or Rename Files

Moves or renames files.

```bash
mv file.txt backup/
```

---

## 10. `rm` – Remove Files

Deletes files.

```bash
rm file.txt
```

---

# Editing Files

## 11. `vim` – Open and Edit Files

Open a file using Vim editor.

```bash
vim file.txt
```

### Save and Exit in Vim

1. Press `ESC`
2. Type:

```bash
:wq
```

3. Press `Enter`

# Package Management Commands

## 12. `sudo apt update`

Refreshes the package list from repositories.

```bash
sudo apt update
```

---

## 13. `sudo apt upgrade`

Installs available software updates.

```bash
sudo apt upgrade
```

# File Permission Commands

## 14. `chmod` – Change File Permissions

```bash
chmod 700 file.txt
```

### Permission Numbers

| Number | Permission |
|---|---|
| 0 | No permission |
| 1 | Execute |
| 2 | Write |
| 4 | Read |
| 7 | Read + Write + Execute |

---

## 15. `chown` – Change File Ownership

```bash
chown user file.txt
```

---

# Service Management Commands

## 31. `systemctl`

Manage Linux services.

### Start Service

```bash
systemctl start nginx
```

### Check Status

```bash
systemctl status nginx
```

### Stop Service

```bash
systemctl stop nginx
```
